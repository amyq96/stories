## Remove duplicate SR Numbers

These are duplicate versions of the same calls that mostly got recategorized as different violations such as ventilation, electrical, plumbing etc.

```python
# identify 249 rows of duplicates
duplicates_mask = df.duplicated(subset=['SR Number'],keep=False)
df[duplicates_mask]
```

```python
# remove duplicates by keeping first instance only in df
df.drop_duplicates(subset=['SR Number'], keep='first', inplace=True)
```

```python
# export deduplicated dataframe
df.to_csv('../data/processed/cleaned_complaints_deduplicated.csv')
```
