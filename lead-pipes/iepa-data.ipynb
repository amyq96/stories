```python
import pandas as pd

df = pd.read_excel('source/Chicago Final Material Inventory.xlsx', skiprows=range(0,5))
```

```python
df.head()
```

```python
# drop extra rows at top
df = df.drop(index=range(1,11))
```

```python
df.head()
```

```python
# make first row the col names
df = df.rename(columns=df.iloc[0])
```

```python
df.head()
```

```python
df = df.drop(index=0)
```

```python
df.head()
```

```python
# extract zip 
df['zipcode'] = df['Service Address'].str.extract(r'(\d{5})')
```

```python
df.tail()
```

```python
df.columns
```

```python
len(df)
```

```python
# filter df for just addresses with confirmed lead (?) in their entire line
# should be 150,750 based on top of file
print(df.groupby('Classification for Entire Serivice Line').size())
lead_df = df[df['Classification for Entire Serivice Line'] == 'L'].copy()
```

```python
# what are the zip codes with the most confirmed lead service lines?
lead_zips = lead_df.groupby('zipcode').size().reset_index(name='count').sort_values('count', ascending=False)
lead_zips
```

```python
# import cook zip code geojson
import geopandas as gpd
cook_map = gpd.read_file('../geographies/Cook_County_Digital_Equity_Score_Zip_Code.geojson')
```

```python
cook_map.plot()
```

```python
cook_map.head()
```

```python
cook_map.info()
```

```python
# merge lead zips to gdf
lead_zips_gdf = cook_map.merge(lead_zips, left_on='zip', right_on='zipcode')
```

```python
lead_zips_gdf.head()
```

```python
# plot 
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

lead_zips_gdf.plot(ax=ax, column='count', cmap='Purples', legend=True)
ax.set_title('Count of addresses w/ confirmed lead service lines')
ax.axis('off')
plt.tight_layout() # adjust layout
plt.show()
```

```python
# what about GRR, unknown, or not lead?

# pivot df by zipcode and entire service line classifcation
line_class = pd.pivot_table(df,
                           index='zipcode',
                           columns='Classification for Entire Serivice Line',
                           values='Service Address',
                           aggfunc='count').reset_index()
# check
print(line_class['L'].sum())

line_class.tail()
```

```python
# merge lead zips to gdf
line_class_gdf = cook_map.merge(line_class, left_on='zip', right_on='zipcode')
```

```python
# Create a figure with four subplots side by side
fig, ax = plt.subplots(1, 4, figsize=(15, 5))

# Plot on the first subplot
line_class_gdf.plot(ax=ax[0], column='L', cmap='Purples', legend=True)
ax[0].set_title('Lead SLs')

# Plot on the second subplot
line_class_gdf.plot(ax=ax[1], column='NL', cmap='Purples', legend=True)
ax[1].set_title('Non-lead SLs')

# Plot on the third subplot
line_class_gdf.plot(ax=ax[2], column='U', cmap='Purples', legend=True)
ax[2].set_title('Unknown SLs')

# Plot on the fourth subplot
line_class_gdf.plot(ax=ax[3], column='GRR', cmap='Purples', legend=True)
ax[3].set_title('GRR SLs')

ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
ax[3].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# consistent axes

# Create a figure with four subplots side by side
fig, ax = plt.subplots(1, 4, figsize=(15, 5))

vmin, vmax = line_class_gdf[['L', 'NL', 'U', 'GRR']].min().min(), line_class_gdf[['L', 'NL', 'U', 'GRR']].max().max()

# Plot on the first subplot
line_class_gdf.plot(ax=ax[0], column='L', cmap='Purples', legend=True, vmin=vmin, vmax=vmax)
ax[0].set_title('Lead SLs')

# Plot on the second subplot
line_class_gdf.plot(ax=ax[1], column='NL', cmap='Purples', legend=True, vmin=vmin, vmax=vmax)
ax[1].set_title('Non-lead SLs')

# Plot on the third subplot
line_class_gdf.plot(ax=ax[2], column='U', cmap='Purples', legend=True, vmin=vmin, vmax=vmax)
ax[2].set_title('Unknown SLs')

# Plot on the fourth subplot
line_class_gdf.plot(ax=ax[3], column='GRR', cmap='Purples', legend=True, vmin=vmin, vmax=vmax)
ax[3].set_title('GRR SLs')

ax[0].axis('off')
ax[1].axis('off')
ax[2].axis('off')
ax[3].axis('off')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
