```python
# remap landlord names

import csv
landlords = [x for x in csv.DictReader(open('manual/Prolific Plaintiffs Dictionary.csv'))]

lookup = {}

for landlord in landlords:
    if landlord['Plaintiff Name'] not in lookup:
        lookup[landlord['Plaintiff Name']] = []
    lookup[landlord['Plaintiff Name']] = landlord['Cleaned Name']
    
# map based on lookup dict to a ne col
df['prolific_plaintiffs']= df['Plaintiff Name'].map(lookup)
```
