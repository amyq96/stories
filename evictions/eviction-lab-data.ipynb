```python
import pandas as pd
```

```python
df = pd.read_csv('source/all_sites_monthly_2020_2021_eviction_lab.csv')
```

```python
df.head()
```

```python
# create a year column
df['month'] = pd.to_datetime(df['month'])
df['year'] = df['month'].dt.year
```

```python
# pivot by filings_2020 by year by city
filings_by_city = pd.pivot_table(df, values='filings_2020', index='city', columns='year', aggfunc='sum').reset_index()
```

```python
# rename cols
columns = ['city','2020_filings', '2021_filings', '2022_filings', '2023_filings']
filings_by_city.columns = columns
```

```python
filings_by_city
```

```python
# pivot historical avg filings by year by city
hist_filings_by_city = pd.pivot_table(df, values='filings_avg', index='city', columns='year', aggfunc='sum').reset_index()
```

```python
# rename cols
columns = ['city','2020_hist_filings', '2021_hist_filings', '2022_hist_filings', '2023_hist_filings']
hist_filings_by_city.columns = columns
```

```python
# typical filings in a given year
hist_filings_by_city
```

```python
# merge by city
change_by_city = filings_by_city.merge(hist_filings_by_city, on='city')
```

```python
change_by_city.head()
```

```python
# calculate change and percent change between 2022 and filings_avg
change_by_city['2022_change'] = change_by_city['2022_filings'] - change_by_city['2022_hist_filings']
change_by_city['2022_pct_change'] = (change_by_city['2022_change']/change_by_city['2022_hist_filings'])*100
```

```python
# gut check: https://evictionlab.org/ets-report-2022/
change_by_city.sort_values('2022_pct_change')
```

```python
# create a df for flourish
flourish_df = change_by_city.copy()
flourish_df = flourish_df[['city', '2022_filings','2022_hist_filings', '2022_change','2022_pct_change']]
```
