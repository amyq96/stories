```python
# community by estimated type for all parcels
by_comm = pd.pivot_table(df,
              index='community',
              columns='estimated_type',
              values='pin',
              aggfunc='count')

by_comm.reset_index().to_csv('output/community_by_est_type.csv', index=False)
```

```python
# community by estiamted type for multifamily
by_comm_multi = pd.pivot_table(df[df['major_class_type'] == 'Multi-Family'],
              index='community',
              columns='estimated_type',
              values='pin',
              aggfunc='count')

by_comm_multi.reset_index().to_csv('output/community_by_est_type_multi.csv', index=False)
```

```python
absentee_out_of_state = df[(df['estimated_type'] == 'Estimated Absentee Owner') & (df['mailing_state'] != 'IL')].copy()

# community by estimated type for all parcels
by_comm_absentee_out_of_state = pd.pivot_table(absentee_out_of_state,
              index='community',
              columns='estimated_type',
              values='pin',
              aggfunc='count')

by_comm_absentee_out_of_state.to_csv('output/absentee_out_of_state_by_comm.csv')
```

```python
# community by estiamted type for Two to six residential apartments, any age
# one of the most common mutli family classses
by_comm_two_six = pd.pivot_table(df[df['class_desc'] == 'Two to six residential apartments, any age'],
              index='community',
              columns='estimated_type',
              values='pin',
              aggfunc='count')

by_comm_two_six.reset_index().to_csv('output/community_by_est_type_two_six.csv', index=False)
```
