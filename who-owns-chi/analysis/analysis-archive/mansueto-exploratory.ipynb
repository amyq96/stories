This notebook loads the raw parquet file data. It only looks at 2006 and 2022 by community area and census tract to determine the count of parcels that match the "is_org" (see Divij's email for exact definitions TK) and "is_homeowner" criteria. 

It outputs dataframes for 2006 and 2022.

# split by category

```python
# filter for residential
df22_r = df22.loc[
    (df22['major_class_code'].isin(['2','3']))] # only counts residential  

df06_r = df06.loc[
    (df06['major_class_code'].isin(['2','3']))]
```

```python
# count of unique parcel pins by comm area for residential 
df22_r_munci = df22_r.groupby('munci_comm')['pin'].count().reset_index(name='res_22')
df06_r_munci = df06_r.groupby('munci_comm')['pin'].count().reset_index(name='res_06')
df06_r_munci
```

```python
# count of unique parcel pins by TRACT for residential 
df22_r_tract = df22_r.groupby('tract_GEOID')['pin'].count().reset_index(name='res_22')
df06_r_tract = df06_r.groupby('tract_GEOID')['pin'].count().reset_index(name='res_06')
df06_r_tract.head()
```

```python
# filter for residential is_org
df22_org = df22.loc[
    (df22['major_class_code'].isin(['2','3'])) & # only counts residential    
    (df22['is_org'] == 1.0)] # only counts if parcel is defined as corporate

df06_org = df06.loc[
    (df06['major_class_code'].isin(['2','3'])) & # only counts residential    
    (df06['is_org'] == 1.0)] # only counts if parcel is defined as corporate

df06_org.head()
```

```python
# count of unique parcel pins by comm area for residential corporate
df22_org_munci = df22_org.groupby('munci_comm')['pin'].count().reset_index(name='is_org_22')
df06_org_munci = df06_org.groupby('munci_comm')['pin'].count().reset_index(name='is_org_06')
df06_org_munci
```

```python
# count of unique parcel pins by comm area for residential corporate
df22_org_tract = df22_org.groupby('tract_GEOID')['pin'].count().reset_index(name='is_org_22')
df06_org_tract = df06_org.groupby('tract_GEOID')['pin'].count().reset_index(name='is_org_06')
df06_org_tract
```

```python
# filter for residential is not a homeowner

df22_home = df22.loc[
    (df22['major_class_code'].isin(['2','3'])) & # only counts residential    
    (df22['is_homeowner'] == 0.0)] # only counts if parcel is defined as not taking homeowners exemption + different mailing addresses 

df06_home = df06.loc[
    (df06['major_class_code'].isin(['2','3'])) & # only counts residential    
    (df06['is_homeowner'] == 0.0)] # only counts if parcel is defined as not taking homeowners exemption + different mailing addresses   
```

```python
# count of unique parcel pins by comm area for residential corporate
df22_home_munci = df22_home.groupby('munci_comm')['pin'].count().reset_index(name='is_not_homeowner_22')
df06_home_munci = df06_home.groupby('munci_comm')['pin'].count().reset_index(name='is_not_homeowner_06')
df06_home_munci
```

```python
# count of unique parcel pins by comm area for residential corporate
df22_home_tract = df22_home.groupby('tract_GEOID')['pin'].count().reset_index(name='is_not_homeowner_22')
df06_home_tract = df06_home.groupby('tract_GEOID')['pin'].count().reset_index(name='is_not_homeowner_06')
df06_home_tract
```

# combine everything

```python
# community area + municipality

# 2006
munci_comm_grouped = pd.merge(df06_r_munci, df06_org_munci, on='munci_comm')
munci_comm_grouped = pd.merge(munci_comm_grouped, df06_home_munci, on='munci_comm')

# 2022
munci_comm_grouped = pd.merge(munci_comm_grouped, df22_r_munci, on='munci_comm')
munci_comm_grouped = pd.merge(munci_comm_grouped, df22_org_munci, on='munci_comm')
munci_comm_grouped = pd.merge(munci_comm_grouped, df22_home_munci, on='munci_comm')
```

```python
munci_comm_grouped.head()
```

```python
# export
munci_comm_grouped.to_csv('output/mansueto_residential_by_munci_comm_2006_2022.csv', index=False)
```

```python
# tract - which tract year?

# 2006
tract_comm_grouped = pd.merge(df06_r_tract, df06_org_tract, on='tract_GEOID')
tract_comm_grouped = pd.merge(tract_comm_grouped, df06_home_tract, on='tract_GEOID')

# 2022
tract_comm_grouped = pd.merge(tract_comm_grouped, df22_r_tract, on='tract_GEOID')
tract_comm_grouped = pd.merge(tract_comm_grouped, df22_org_tract, on='tract_GEOID')
tract_comm_grouped = pd.merge(tract_comm_grouped, df22_home_tract, on='tract_GEOID')
```

```python
tract_comm_grouped
```

```python
tract_comm_grouped.to_csv('output/mansueto_residential_by_tract_2006_2022.csv', index=False)
```
